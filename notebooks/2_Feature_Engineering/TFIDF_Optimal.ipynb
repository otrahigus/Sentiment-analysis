```python
import pandas as pd
import numpy as np

data_train = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/data project 1 AI Indonesia/data_train.csv')
data_test = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/data project 1 AI Indonesia/data_test.csv')
```

```python
sample_tweets = data_train['clean_tweet'].sample(10, random_state=42).to_list()
for i, clean_tweet in enumerate(sample_tweets, 1):
  print(f"Tweet ke-{i}: {clean_tweet}")
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
y_train_enc = encoder.fit_transform(data_train['sentimen'])  # ['positif', 'negatif'] → [2, 0]
y_test_enc = encoder.transform(data_test['sentimen'])
```

```python
#TF-IDF dengan Parameter Optimal
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_optimal = TfidfVectorizer(
    max_features=5000,        # Batasi jumlah fitur
    ngram_range=(1, 3),       # Unigram + bigram + trigram
    min_df=2,                  # Minimal muncul di 2 dokumen
    max_df=0.9,                # Abaikan kata yang muncul di >90% dokumen
    sublinear_tf=True          # Gunakan 1 + log(tf)
)
x_train_tfidf = tfidf_optimal.fit_transform(data_train['clean_tweet'])
x_test_tfidf = tfidf_optimal.transform(data_test['clean_tweet'])

print("📋 Parameter TF-IDF:")
print(f"   • max_features: {tfidf_optimal.max_features}")
print(f"   • ngram_range: {tfidf_optimal.ngram_range}")
print(f"   • min_df: {tfidf_optimal.min_df}")
print(f"   • max_df: {tfidf_optimal.max_df}")
print(f"   • sublinear_tf: {tfidf_optimal.sublinear_tf}")
```

```python
feature_names = tfidf_optimal.get_feature_names_out()
print(f"\n   📝 Contoh fitur (10 pertama):")
print(f"   {feature_names[:10]}")
```

```python
#Menyimpan matrik hasil tf idf dan label encoder untuk efesiensi

import joblib
from scipy.sparse import save_npz, load_npz

alldata_tfidf_optimal = {
    'x_train_tfidf_optimal': x_train_tfidf,
    'x_test_tfidf_optimal': x_test_tfidf,
    'encoder': encoder,
    'tfidf_optimal': tfidf_optimal,
    'y_train_enc': y_train_enc,
    'y_test_enc': y_test_enc,
    'feature_names': feature_names
}

joblib.dump(alldata_tfidf_optimal, '/content/drive/MyDrive/Colab Notebooks/data project 1 AI Indonesia/Individual Project/TF-IDF&LogisticRegression/data/alldata_tfidf_optimal.pkl')
```
