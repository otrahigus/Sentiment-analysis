```python
import pandas as pd
import numpy as np

data_train = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/data project 1 AI Indonesia/data_train.csv')
data_test = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/data project 1 AI Indonesia/data_test.csv')
```

```python
data_train.head()
```

```python
data_test.head()
```

```python
sample_tweets = data_train['clean_tweet'].sample(10, random_state=42).to_list()
for i, clean_tweet in enumerate(sample_tweets, 1):
  print(f"Tweet ke-{i}: {clean_tweet}")
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
y_train_enc = encoder.fit_transform(data_train['sentimen'])  # ['positif', 'negatif'] → [2, 0]
y_test_enc = encoder.transform(data_test['sentimen'])
```

```python
#TF-IDF Defaults
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_defaults = TfidfVectorizer()
x_train_tfidf = tfidf_defaults.fit_transform(data_train['clean_tweet'])
x_test_tfidf = tfidf_defaults.transform(data_test['clean_tweet'])
```

```python
feature_names = tfidf_defaults.get_feature_names_out()
print(f"\n   📝 Contoh fitur (10 pertama):")
print(f"   {feature_names[:10]}")
```

```python
#Menyimpan matrik hasil tf idf dan label encoder untuk efesiensi

import joblib
from scipy.sparse import save_npz, load_npz

all_data = {
    'x_train_tfidf': x_train_tfidf,
    'x_test_tfidf': x_test_tfidf,
    'encoder': encoder,
    'tfidf_defaults': tfidf_defaults,
    'y_train_enc': y_train_enc,
    'y_test_enc': y_test_enc,
    'feature_names': feature_names
}

joblib.dump(all_data, '/content/drive/MyDrive/Colab Notebooks/data project 1 AI Indonesia/Individual Project/TF-IDF&LogisticRegression/data/all_data.pkl')
```
